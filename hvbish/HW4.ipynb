```python
df['had_aids'] = df[cb.variable[76]]=='Yes'
df['aids_death'] = (df['gs_text34']=='AIDS')
df['not_aids_death'] = 1 - df['aids_death']
df.pivot_table(values=['aids_death','not_aids_death'],index='had_aids',aggfunc=sum)
```

```python
for i in df.groupby('site'):
    print('site ', i[0])
    print pd.crosstab(i[1]['had_aids'],i[1]['aids_death'])
    print('')
```

```python
dab = pd.crosstab(i[1]['had_aids'],i[1]['aids_death'])
dab.apply(lambda r: r / r.sum()*100, axis=1)
```

```python
#for cases where the "had AIDS" question was endorsed, about 93% of those deaths were truly due to AIDS
```
